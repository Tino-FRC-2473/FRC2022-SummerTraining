# **ball contour function**

```python
RED_MEAN = [0.43410833, -0.07345035, 0.27004727]
RED_STD = [0.1053285, 0.01958057, 0.06364335]

BLUE_MEAN = [0.33700416498038976, 0.13068530477054924, -0.12186020618901487]
BLUE_STD = [0.09878337074515638, 0.03583767932011143, 0.03559789543653717]
```

```python
imageheight = 0

def red_filter(name):
  im = bgr_to_rgb(cv2.imread(name))

  imgyuv = skimage.color.rgb2yuv(im)

  scale = 2
  low_threshold = np.array([0.0,
                            RED_MEAN[1] - scale * RED_STD[1],
                            RED_MEAN[2] - scale * RED_STD[2]])

  high_threshold = np.array([1.0,
                            RED_MEAN[1] + scale * RED_STD[1],
                            RED_MEAN[2] + scale * RED_STD[2]])

  background_mask = np.all((imgyuv > low_threshold) * (imgyuv < high_threshold), axis=2)
  background_mask = np.where(background_mask, 1, 0)
  background_mask = np.expand_dims(background_mask, 2)

  im = im*background_mask
  cv2.imwrite("mid-code-red.jpg", im)

  contour("mid-code-red.jpg")

def blue_filter(name):
  im = bgr_to_rgb(cv2.imread(name))

  imgyuv = skimage.color.rgb2yuv(im)

  scale = 2
  low_threshold = np.array([0.0,
                            BLUE_MEAN[1] - scale * BLUE_STD[1],
                            BLUE_MEAN[2] - scale * BLUE_STD[2]])

  high_threshold = np.array([1.0,
                            BLUE_MEAN[1] + scale * BLUE_STD[1],
                            BLUE_MEAN[2] + scale * BLUE_STD[2]])

  background_mask = np.all((imgyuv > low_threshold) * (imgyuv < high_threshold), axis=2)
  background_mask = np.where(background_mask, 1, 0)
  background_mask = np.expand_dims(background_mask, 2)

  im = im*background_mask
  cv2.imwrite("mid-code-blue.jpg", im)

  contour("mid-code-blue.jpg")

def contour(name):
  img = cv2.imread(name)

  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  img = cv2.blur(img, (10, 10))

  imgray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

  ret, thresh = cv2.threshold(imgray, 50, 160, 0)

  contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

  img = skimage.color.yuv2rgb(img)

  c = find_largest_square_contour(contours)
  x, y, w, h = cv2.boundingRect(c)
  imageheight = h
  cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 3)

  plt.imshow(img)
  plt.show()

def find_largest_square_contour(contours):
  largest = contours[0]
  for c in contours:
    if (cv2.contourArea(c) > cv2.contourArea(largest)): 
      x,y,w,h = cv2.boundingRect(c);
      if (w/h > 0.7 and w/h < 1.3):
        largest = c
  return largest
```

```python
LIMELIGHT_RED_MEAN = [0.21315404333333335, -0.10262768, 0.41609857]
LIMELIGHT_RED_STD = [0.035980533333333335, 0.0175894, 0.06752502]

def red_limelight(name):
  red_processing_l(name)
  distance("mid-code-red.jpg")

def red_processing_l(name):
  im = bgr_to_rgb(cv2.imread(name))

  imgyuv = skimage.color.rgb2yuv(im)

  scale = 4
  low_threshold = np.array([0.0,
                            LIMELIGHT_RED_MEAN[1] - scale * LIMELIGHT_RED_STD[1],
                            LIMELIGHT_RED_MEAN[2] - scale * LIMELIGHT_RED_STD[2]])

  high_threshold = np.array([1.0,
                            LIMELIGHT_RED_MEAN[1] + scale * LIMELIGHT_RED_STD[1],
                            LIMELIGHT_RED_MEAN[2] + scale * LIMELIGHT_RED_STD[2]])

  background_mask = np.all((imgyuv > low_threshold) * (imgyuv < high_threshold), axis=2)
  background_mask = np.where(background_mask, 1, 0)
  background_mask = np.expand_dims(background_mask, 2)

  im = im*background_mask
  cv2.imwrite("mid-code-red.jpg", im)

def distance(name):
  img = cv2.imread(name)

  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  img = cv2.blur(img, (10, 10))

  imgray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

  ret, thresh = cv2.threshold(imgray, 10, 240, 0)

  contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

  img = skimage.color.yuv2rgb(img)

  c = find_largest_square_contour(contours)
  x, y, w, h = cv2.boundingRect(c)
  imgh = h*2.54/96
  dist = (13.39*24.13/imgh)/2.54/12
  print(dist, "feet away")

  cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 3)

  plt.imshow(img)
  plt.show()

def find_largest_square_contour(contours):
  largest = contours[0]
  for c in contours:
    if (cv2.contourArea(c) > cv2.contourArea(largest)): 
      x,y,w,h = cv2.boundingRect(c);
      if (w/h > 0.7 and w/h < 1.3):
        largest = c
  return largest
```

i guess to explain the contour code:

*   start off with an image where you know where the reflective tape is
*   determine the average yuv color and standard deviation for that
*   set the low threshold to the average - the standard deviation and the high threshold to the average + the standard deviation
*   layer that mask over the image to find the green parts
*   then run a contour and run calculations if necessary to make sure it's the right one (like width/height ratio, etc)
